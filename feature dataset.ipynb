```python
data = {
    "Category": ["Electronics", "Clothing", "Home Appliances", "Books", "Toys"],
    "Sales": [450, 300, 150, 200, 100]
}

# Create the donut chart
fig = px.pie(
    data_frame=data,
    names="Category",
    values="Sales",
    title="Sales Distribution by Category",
    hole=0.4,  
    color_discrete_sequence=px.colors.sequential.Plasma
)
fig.show()
```

```python
data = {
    "Category": ["All", "All", "Electronics", "Electronics"],
    "Subcategory": ["Electronics", "Clothing", "Phones", "Laptops"],
    "Sales": [1000, 500, 400, 600]
}


fig = px.icicle(
    names=data["Subcategory"],
    parents=data["Category"],
    values=data["Sales"],
    title="Icicle Chart: Sales Distribution",
    color=data["Sales"],
    color_continuous_scale="Blues"
)


fig.show()
```
